# STEP 1: Data Integration
In this notebook we integrate all datasets into one unique datasets to facilitate the analysis.

```python
#import libraries
import pandas as pd
import itertools
```

### 1. Read countries_population.csv file
The file contains the population of different countries from years 2000 to 2019.

```python
df_countries_population = pd.read_csv('INPUT DATA/countries_population.csv')
df_countries_population.rename(columns={'Country Code': 'country_code'}, inplace=True)
df_countries_population.head()
```

### 2. Read countries_continent.csv file
It contains basic information of the countries (continent, code…)

```python
df_countries_continent = pd.read_csv('INPUT DATA/countries_continent.csv', encoding = "utf-8")
df_countries_continent = df_countries_continent.rename({'country_code': 'num_code'}, axis=1)
df_countries_continent = df_countries_continent.rename({'code_3': 'country_code'}, axis=1)
df_countries_continent.head()
```

### 3. Read countries_deathrate.csv file
The file contains the deathrates of different countries from years 2010 to 2019.

```python
df_countries_deathrate = pd.read_csv('INPUT DATA/countries_deathrate.csv', encoding = "utf-8")
df_countries_deathrate.rename(columns={'Country Code': 'country_code'}, inplace=True)
df_countries_deathrate.head()
```

### 4. Read countries_gdp.csv 
It contains the gross domestic product (GDP) value of different countries.

```python
df_countries_gdp = pd.read_csv('INPUT DATA/countries_gdp.csv', error_bad_lines=False)
df_countries_gdp.rename(columns={'Country Code': 'country_code'}, inplace=True)
df_countries_gdp.head()
```

### 5. Read active_HIV_cases_children.csv file
The file contains the active HIV cases of children (under 15 years) in different countries from years 1990 to 2018.

```python
df_active_HIV_cases_children = pd.read_csv('INPUT DATA/active_HIV_cases_children.csv', error_bad_lines=False)
df_active_HIV_cases_children.rename(columns={'Country': 'country_name'}, inplace=True)
df_active_HIV_cases_children.head()
```

### 6. Read active_HIV_cases_male_adults.csv file
It contains active HIV cases of male adults from the different countries from years 1990 to 2018.

```python
df_active_HIV_cases_male_adults = pd.read_csv('INPUT DATA/active_HIV_cases_male_adults.csv', error_bad_lines=False)
df_active_HIV_cases_male_adults.rename(columns={'Country': 'country_name'}, inplace=True)
df_active_HIV_cases_male_adults.head()
```

### 7. Read active_HIV_cases_female_adults.csv file
It contains active HIV cases of female adults from the different countries from years 1990 to 2018.

```python
df_active_HIV_cases_female_adults = pd.read_csv('INPUT DATA/active_HIV_cases_female_adults.csv', error_bad_lines=False)
df_active_HIV_cases_female_adults.rename(columns={'Country': 'country_name'}, inplace=True)
df_active_HIV_cases_female_adults.head()
```

### 8. Read aids_deaths_children.csv file
It contains active AIDS deaths of childrens from the different countries from years 1990 to 2018.

```python
df_aids_deaths_children = pd.read_csv('INPUT DATA/aids_deaths_children.csv', error_bad_lines=False)
df_aids_deaths_children.rename(columns={'Country': 'country_name'}, inplace=True)
df_aids_deaths_children.head()
```

### 9. Read aids_deaths_male_adults.csv file
It contains active AIDS deaths of male adults from the different countries from years 1990 to 2018.

```python
df_aids_deaths_male_adults = pd.read_csv('INPUT DATA/aids_deaths_male_adults.csv', error_bad_lines=False)
df_aids_deaths_male_adults.rename(columns={'Country': 'country_name'}, inplace=True)
df_aids_deaths_male_adults.head()
```

### 10. Read aids_deaths_female_adults.csv file
It contains active AIDS deaths of female adults from the different countries from years 1990 to 2018.

```python
df_aids_deaths_female_adults = pd.read_csv('INPUT DATA/aids_deaths_female_adults.csv', error_bad_lines=False)
df_aids_deaths_female_adults.rename(columns={'Country': 'country_name'}, inplace=True)
df_aids_deaths_female_adults.head()
```

### 11. Join the datasets

#### Join the countries with the corresponding codes

```python
#get countries from datasets
countries = df_countries_population['Country Name']
codes = df_countries_population['country_code']
df_countries = pd.concat([countries,codes], axis=1)
df_countries.columns = ['country_name','country_code']
df_countries.head()
```

```python
#join the countries, country code and other info
df_countries = df_countries.merge(df_countries_continent, on='country_code', how='left')
#remove inncesary columns
df_countries = df_countries.drop(columns=['country','code_2','num_code','iso_3166_2','region_code','sub_region_code'])
df_countries.head()
```

```python
years = ['2010','2011','2012','2013','2014','2015','2016','2017','2018']
cols = ['year','country_name','country_code','continent','sub_region']
country_codes = pd.Series(df_countries_population['country_code'].dropna(axis=0))
df_all = pd.DataFrame(columns=cols)

for y in years :
    year_series = pd.Series([y] * len(df_countries))
    df = pd.concat([year_series, country_codes], axis=1)
    df.columns = ['year','country_code']
    df = df.merge(df_countries, on='country_code', how='left').dropna(axis=0)
    df_all = df_all.append(df, ignore_index=True)
df_all
```

#### Create a function to join the datasets

```python
def join_data(org_df, join_df, name, aids) :
    new_df = pd.DataFrame()
    years = ['2010','2011','2012','2013','2014','2015','2016','2017','2018']
    for y in years :
        if aids == False :
            col = [s for s in join_df.columns.tolist() if y in s]
            merge_cols = ['country_code', col[0]]
        else :
            col = [' '+y]
            merge_cols = ['country_name', col[0]]
        df2 = df_all[df_all['year'] == y]
        cols = df2.columns.tolist()
        cols.append(name)
        if aids == False :
            df2 = df2.merge(join_df[merge_cols], on='country_code', how='left')
        else :
            df2 = df2.merge(join_df[merge_cols], on='country_name', how='left')
        df2.columns = cols
        new_df = new_df.append(df2,ignore_index=True)
    return new_df
```

#### Join the population of each country from years 2010 to 2018

```python
df_all = join_data(df_all, df_countries_population, 'population', aids=False)
df_all
```

#### Join the deathrate of each country from years 2010 to 2018

```python
df_all = join_data(df_all, df_countries_deathrate, 'deathrate', aids=False)
df_all
```

#### Join the gdp of each country from years 2010 to 2018

```python
df_all = join_data(df_all, df_countries_gdp, 'gdp_per_capita', aids=False)
df_all
```

#### Join the active hiv cases of each country from years 2010 to 2018

```python
df_all = join_data(df_all, df_active_HIV_cases_children, 'active_HIV_cases_children', aids=True)
df_all = join_data(df_all, df_active_HIV_cases_male_adults, 'active_HIV_cases_male_adults', aids=True)
df_all = join_data(df_all, df_active_HIV_cases_female_adults, 'active_HIV_cases_female_adults', aids=True)
df_all
```

#### Join the active AIDS deaths of each country from years 2010 to 2018

```python
df_all = join_data(df_all, df_aids_deaths_children, 'AIDS_deaths_children', aids=True)
df_all = join_data(df_all, df_aids_deaths_male_adults, 'AIDS_deaths_male_adults', aids=True)
df_all = join_data(df_all, df_aids_deaths_female_adults, 'AIDS_deaths_female_adults', aids=True)
df_all
```

### 12. Save the created dataset

```python
df_all.to_csv('FINAL DATA/AIDS_GDP_WORLD_DATASET.csv', index=False)
```
