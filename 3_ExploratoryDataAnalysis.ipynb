```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Read the dataset

```python
df = pd.read_csv('AIDS_GDP_CLEANED.csv')
df
```

### 2. Analyse descriptive statistics

```python
df.describe(include='all')
```

### 3. Analyse information of the countries

```python
# grouping results
cols = ['country_name','continent','sub_region']
df_grouped = df[cols]
grouped_test1 = df_grouped.groupby(['country_name','sub_region'], as_index=False)
grouped_test1.describe()
```

### 4. Histograms, density plots and boxplots for numerical attributes

```python
num_cols = ['population', 'deathrate', 'gdp_per_capita','active_HIV_cases_children', 'active_HIV_cases_male_adults','active_HIV_cases_female_adults', 'AIDS_deaths_children',
            'AIDS_deaths_male_adults', 'AIDS_deaths_female_adults']
len(num_cols)
```

```python
#histograms
fig, axs = plt.subplots(3,3, figsize=(13,7.5))
idx = {0:[0,0], 1:[0,1], 2:[0,2], 3:[1,0], 4:[1,1], 5:[1,2], 6:[2,0], 7:[2,1], 8:[2,2]}

for i in range(0,len(num_cols)) :
    ax = axs[idx[i][0],idx[i][1]]
    ax.hist(df[num_cols[i]].values)
    ax.set_title(num_cols[i])
    
fig.tight_layout()
```

```python
#density plots
fig, axs = plt.subplots(3,3, figsize=(13,7.5))
idx = {0:[0,0], 1:[0,1], 2:[0,2], 3:[1,0], 4:[1,1], 5:[1,2], 6:[2,0], 7:[2,1], 8:[2,2]}

for i in range(0,len(num_cols)) :
    ax = axs[idx[i][0],idx[i][1]]
    sns.distplot(df[num_cols[i]].values, hist=False, ax=ax)
    ax.set_title(num_cols[i])
    
fig.tight_layout()
```

```python
#boxplots
fig, axs = plt.subplots(3,3, figsize=(13,7.5))
idx = {0:[0,0], 1:[0,1], 2:[0,2], 3:[1,0], 4:[1,1], 5:[1,2], 6:[2,0], 7:[2,1], 8:[2,2]}

for i in range(0,len(num_cols)) :
    ax = axs[idx[i][0],idx[i][1]]
    ax.boxplot(df[num_cols[i]].values)
    ax.set_title(num_cols[i])
    
fig.tight_layout()
```

### 5. Correlations between numerical attributes

```python
cors = df[num_cols].corr(method='pearson')

fig, ax = plt.subplots(figsize=(15,5))
sns.heatmap(cors, mask=np.triu(np.ones_like(cors, dtype=bool)), cmap='Blues', square=True, ax=ax, linewidth=.5)
```
