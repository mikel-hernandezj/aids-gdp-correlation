# 6. Population groups analysis by years
In this notebook an analysis has been performed to analyse the differences between the three groups (children, male adults and female adults)

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import scipy.optimize as op
```

```python
df = pd.read_csv('AIDS_GDP_CLEANED.csv')
df.head()
```

### 1. Group the dataset by year

```python
df_grouped = df.groupby(['year'],as_index=False).mean()
df_grouped
```

### 2. Analyse the active HIV cases of the three groups (children, male adults and female adults)

```python
hiv_cases_data = df_grouped[['active_HIV_cases_children','active_HIV_cases_male_adults','active_HIV_cases_female_adults']].values
hiv_cases_data.shape
```

```python
plt.boxplot(hiv_cases_data)
plt.title('Boxplots of active HIV cases in the world')
plt.xticks([1,2,3], ['children','male_adults','female_adults'])
plt.show()
```
